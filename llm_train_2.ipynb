```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoModelForCausalLM
```

```python
import torch
```

```python
torch_device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
torch_device
```

```python
model_name = "mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name,pad_token_id=tokenizer.eos_token_id).to(torch_device)
```

```python
inputs = tokenizer("This is a test", return_tensors="pt").to(torch_device)
```

```python
output = model(**inputs).logits.argmax(1)
```

```python
output
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name,pad_token_id=tokenizer.eos_token_id,lables=['pos','net', 'neg']).to(torch_device)
```

```python
inputs = tokenizer("Nice has reported very god performance for the year", return_tensors="pt").to(torch_device)
```

```python
output = model(**inputs).logits
```
